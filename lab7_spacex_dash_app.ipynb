```python
import pandas as pd
spacex_df = pd.read_csv("spacex_launch_dash.csv")
status = list()
for i in spacex_df["class"]:
    if i == 1:
        status.append("Success")
    else:
        status.append("Failure")
spacex_df["status"] = status
spacex_df
```

```python
data = spacex_df.groupby(["Payload Mass (kg)"], as_index=False).mean()
data
```
